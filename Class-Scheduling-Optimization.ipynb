# Descripción del problema
El problema de programación de clases universitarias (University Scheduling Problem) consiste en la asignación óptima de horarios y aulas a un conjunto de cursos y profesores de una universidad. Este problema es muy complejo debido a la gran cantidad de restricciones que deben ser consideradas, por ejemplo, número limitado de aulas disponibles, la disponibilidad de los profesores, el tipo de materia que cada profesor puede dar y la capacidad de los salones.

Algunos recursos necesarios para llevar a cabo la programación de los grupos serán:
1. Datos profesores (nombre, nómina y correo).
2. Disponibilidad de profesores (día y hora).
3. Claves a programar.
4. Número de grupos de cada clave.
5. Cupo máximo en cada grupo (depende del salón). En promedio 30 alumnos en bloques y 35 en materias.
6. Rango de horario en el que se puede programar los grupos.
7. Conteo de carga máxima semestral de cada profesor (10 unidades de carga para cátedra y 14 para planta. El profesor investigador cuenta con una carga     máxima de 7. Puede haber alguna descarga especial para los profesores, ya sea por alguna enfermedad, proyecto especial, etc).
8. Revisar que no hay empalmes (microprogramación).
9. Poner atributos de cada unidad de formación (grupo en inglés, grupo regional, duración, etc).
10. Acomodo de los paquetes (conjunto de bloques y materias de manera uniforme para que el alumno tenga la mínima cantidad de horas muertas entre clases. Los paquetes también deberían ayudar a que los profesores tengan un horario matutino o vespertino).

Lo anteriormente mencionado depende de los pronósticos que se hagan respecto al número de alumnos por inscribirse. Sin embargo, esto no es completamente confiable debido a los cambios de último momento. Para este caso asumimos que el número de clases y alumnos registrados por clase son conocidos.

El objetivo de este problema es programar las clases universitarias del Instituto Tecnológico de Estudios Superiores Monterrey Campus CM de la manera más eficiente posible, asegurándose de que todas las restricciones sean cumplidas y minimizando los costos y el tiempo de programación. Dentro de esto consideramos los cambios que se quieren realizar dentro de la solución de los problemas principales, los cuales son, una programación sistematizada, que los archivos se liguen de manera eficiente y contar con una lista de espera de profesores.

Sin embargo, como ya se mencionó anteriormente, el conjunto de restricciones y factores que rodean esta situación, hace que esto se trate de un problema muy complejo, el cual no será posible resolver en cuestión de 5 semanas. Aunque nosotros tenemos el compromiso de presentar una propuesta que aporte de manera positiva a la solución de esta problemática.

# Formulación del problema

**Variables de decisión**

\begin{equation*} X_{ijk} = \begin{cases} 1, & \text{si el curso $i$ es impartido por el profesor $j$, en el horario $k$ } \\ 0, & \text{en caso contrario} \end{cases} \end{equation*}

donde:
* $i = {[1, 221]}$
* $j = {[1,70]}$
* $k = {[1,30]}$

\begin{equation*} Y_{i} = \begin{cases} 1, & \text{Si el profesor $i$ esta impartiendo un curso } \\ 0, & \text{en caso contrario} \end{cases} \end{equation*}

donde:
* $i = {[1, 70]}$

**Función Objetivo**

\begin{equation*}
\text{Min}(Z) = \sum \limits_{i\in profesor} Y_{i} 
\end{equation*}

**Restricciones**

En está restricción verificamos que los bloques cumplan  con las horas que se les asignen. En este caso en la posición de subíndice $i$ puede valores en un rango de entre 0 y el número de materias que se quieran asignar, entonces en total tendríamos $i$ restricciones, hay que considerar que se está multiplicando por 10 debido a que tenemos materias con duración de 5 semanas y cada turno dura 2 hrs. $M_{i}$ será un arreglo que contenga las horas de cada materia según corresponda.

\begin{equation*}
10\sum\limits_{j\in profesor}\sum\limits_{k\in  horario} X_{i,j,k} = M_{i}\quad \forall \thinspace i \in materia
\end{equation*}

En esta restricción verificamos que el maestro no sobrepase su carga máxima semestral. Para esto necesitaremos de una matriz $C_{j}$ la cual contendrá la carga máxima de cada profesor, aquí el índice $j$ de $X$ y el índice $j$ de $C$ tienen que ser iguales a la hora de realizar esta comparación. Esto es una forma general de representar $j$ (el número de profesores) restricciones que se deben de realizar.

\begin{equation*}
10\sum\limits_{i\in materia}\sum\limits_{k\in horario} X_{i,j,k} \le C_{j}\quad \forall \thinspace j \in profesor
\end{equation*}

Con esta restricción nos aseguramos de que no haya empalmes en los horarios de los profesores. Esto nos daría un total de $j\cdot k$ restricciones, donde $j$ es el número de profesores y $k$ el número de horarios.

\begin{equation*}
\sum\limits_{i\in materia} X_{i,j,k} \leq 1\quad \forall \thinspace j \in profesor, \forall \thinspace k \in horario
\end{equation*}

Con esta restricción nos aseguramos de que el profesor $j$ está impartiendo un materia $i$ en un horario $k$ se vea reflejado en nuestra variable de decisión $Y$. Es una manera de relacionar ambas variables de decisión. Esto nos daría un total de $i\cdot j\cdot k$ restricciones, donde $i$ es el número de materias, $j$ es el número de profesores y $k$ el número de horarios.

\begin{equation*}
X_{ijk} \leq Y_{j} \quad \forall \thinspace i \in materia, \forall \thinspace j \in profesor,\forall \thinspace k \in horario
\end{equation*}

Esta restricción hace que no haya empalmes en el horario en el que se esté impartiendo la materia. Esto nos daría un total de $i\cdot k$ restricciones, donde $i$ es el número de materias y $k$ el número de horarios.
\begin{equation*}
\sum\limits_{j\in profesor} X_{i,j,k} \leq 1\quad \forall \thinspace i \in materia, \forall \thinspace k \in horario
\end{equation*}


# Parametros del modelo

* Datos del profesor (nombre, nómina y correo).

* Claves a programar y su número de grupos.
* Capacidad máxima de alumnos en el salón.
* Rango de horario en el que se pueden programar los grupos.
* Evitar los empalmes entre materias.
* Acomodar el conjunto de bloques y materias de manera uniforme (mínimo de horas muertas entre clases). 
* Atributos de la Unidad de Formación (idioma, tipo de grupo y duración de la UF)
* Cantidad de alumnos a inscribirse 
* ${C}_{j}$ = carga máxima del profesor $j$
* ${M}_{i}$ = número de horas de la materia $i$

# Implementación de Gurobi

```python
#Importamos las librerias necesarias
import pandas as pd
import numpy as np
import gurobipy as gp
from gurobipy import GRB 
```

```python
#Importamos nuestras bases de datos a usar 
clases = "C:/Users/merar/Documents/Proyecto OD/Programacion_clases.xlsx"
profesores = "C:/Users/merar/Documents/Proyecto OD/Profesores.xlsx"
```

```python
df_clases = pd.read_excel(clases)
df_profesores = pd.read_excel(profesores, usecols=[3]) 
df_profesores1=pd.read_excel(profesores)
```

```python
#Creamos nuestro modelo
m = gp.Model("Profesores") 
```

```python
#Valores de cada subíndice con respecto a la variable de decisión X

#Modelo con instancia chica con 0.1s de ejecución
i_values = range(10)
j_values = range(6)
k_values = range(30)

#Modelo con instancia grande con 1.21 min de ejecución
#i_values = range(93)
#j_values = range(70)
#k_values = range(30)
```

```python
#Crear las variables de decisión
X = m.addVars(i_values, j_values, k_values, vtype=GRB.BINARY, name="X")
y = m.addVars(j_values, vtype=GRB.BINARY, name="y")
```

```python
#Se muestran las variables de decisión de Xijk creadas
for i in i_values:
    for j in j_values:
        for k in k_values:
            variable = X[i, j, k]
            print("Variable ({}, {}, {}) = {}".format(i, j, k, variable))
```

```python
#Se muestran las variables de decisión de Yi creadas
for i in j_values:
    variabley = y[i]
    print("Variable ({}) = {}".format(i,variabley))
```

```python
#Definir la función objetivo y las restricciones existentes
m.setObjective(gp.quicksum(y[i] for i in j_values), GRB.MINIMIZE)
```

```python
M_i = df_clases['Horas']
```

```python
#Restricción 1
for i in i_values:
    m.addConstr(10 * gp.quicksum(X[i,j,k] for j in j_values for k in k_values) == M_i[i], name='Restricción 1')
```

```python
df_profesores = df_profesores.iloc[:len(j_values), :]
C_i = df_profesores.to_numpy()
```

```python
#Restriccion 2
for j in j_values:
    m.addConstr(10 * gp.quicksum(X[i,j,k] for i in i_values for k in k_values) <= C_i[j], name='Restricción 2')
```

```python
#Restriccion 3
for j in j_values:
    for k in k_values:
        m.addConstr(gp.quicksum(X[i,j,k] for i in i_values) <= 1, name='Restricción 3')
```

```python
#Restrcción 4
for j in j_values:
    for i in i_values:
        for k in k_values:
            m.addConstr(X[i,j,k] <= y[j], name='Restricción 4')
```

```python
#Restriccion 5
for i in i_values:
    for k in k_values:
        m.addConstr(gp.quicksum(X[i,j,k] for j in j_values) <= 1, name='Restricción 5')
```

```python
m.optimize()
```

```python
Lunes=[0,1,2,3,4,5]
Martes=[6,7,8,9,10,11]
Miercoles=[12,13,14,15,16,17]
Jueves=[18,19,20,21,23,23]
Viernes=[24,25,26,27,28,29]
def asignar_horario(row):
    if (row['Horario'] in Lunes):
        if Lunes.index(row['Horario'])==0:
            return '07:00 - 09:00'
        elif Lunes.index(row['Horario']) == 1:
            return '09:00 - 11:00'
        elif Lunes.index(row['Horario']) == 2:
            return '11:00 - 13:00'
        elif Lunes.index(row['Horario']) == 3:
            return '13:00 - 15:00'
        elif Lunes.index(row['Horario']) == 4:
            return '15:00 - 17:00'
        else:
            return '17:00 - 19:00'

    elif (row['Horario'] in Martes):
        if Martes.index(row['Horario'])==0:
            return '07:00 - 09:00'
        elif Martes.index(row['Horario']) == 1:
            return '09:00 - 11:00'
        elif Martes.index(row['Horario']) == 2:
            return '11:00 - 13:00'
        elif Martes.index(row['Horario']) == 3:
            return '13:00 - 15:00'
        elif Martes.index(row['Horario']) == 4:
            return '15:00 - 17:00'
        else:
            return '17:00 - 19:00'
    
    elif (row['Horario'] in Miercoles):
        if Miercoles.index(row['Horario'])==0:
            return '07:00 - 09:00'
        elif Miercoles.index(row['Horario']) == 1:
            return '09:00 - 11:00'
        elif Miercoles.index(row['Horario']) == 2:
            return '11:00 - 13:00'
        elif Miercoles.index(row['Horario']) == 3:
            return '13:00 - 15:00'
        elif Miercoles.index(row['Horario']) == 4:
            return '15:00 - 17:00'
        else:
            return '17:00 - 19:00'
    
    elif (row['Horario'] in Jueves):
        if Jueves.index(row['Horario'])==0:
            return '07:00 - 09:00'
        elif Jueves.index(row['Horario']) == 1:
            return '09:00 - 11:00'
        elif Jueves.index(row['Horario']) == 2:
            return '11:00 - 13:00'
        elif Jueves.index(row['Horario']) == 3:
            return '13:00 - 15:00'
        elif Jueves.index(row['Horario']) == 4:
            return '15:00 - 17:00'
        else:
            return '17:00 - 19:00'
    
    elif (row['Horario'] in Viernes):
        if Viernes.index(row['Horario'])==0:
            return '07:00 - 09:00'
        elif Viernes.index(row['Horario']) == 1:
            return '09:00 - 11:00'
        elif Viernes.index(row['Horario']) == 2:
            return '11:00 - 13:00'
        elif Viernes.index(row['Horario']) == 3:
            return '13:00 - 15:00'
        elif Viernes.index(row['Horario']) == 4:
            return '15:00 - 17:00'
        else:
            return '17:00 - 19:00'
        
def asignar_materia(row):
    lm=df_clases['Clave'].tolist()
    valor=lm[row['Materia']]
    return valor+'-'+str(row['Materia'])


def asignar_prof(row):
    return 'Profesor '+str(row['Profesor'])

def asignar_salon(row):
    lm=df_clases['Salón'].tolist()
    valor=lm[row['Materia']]
    return str(valor)

def asignar_dia(row):
    if (row['Horario'] in Lunes):
            return 'Lunes'

    elif (row['Horario'] in Martes):
        return ' Martes'
    
    elif (row['Horario'] in Miercoles):
        return "Miercoles"
    
    elif (row['Horario'] in Jueves):
        return "Jueves"
    
    elif (row['Horario'] in Viernes):
        return "Viernes"

df['Hora'] = df.apply(asignar_horario, axis=1)
df['Clave'] = df.apply(asignar_materia, axis=1)
df['Prof'] = df.apply(asignar_prof, axis=1)
df['Salon'] = df.apply(asignar_salon, axis=1)
df['Dia'] = df.apply(asignar_dia, axis=1)
print(df)
```
